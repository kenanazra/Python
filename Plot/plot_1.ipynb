```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_points = np.array([2,5])
y_points = np.array([0,250])
```

```python
plt.plot(x_points,y_points)
plt.show()
```

```python
plt.plot(x_points,y_points,'*')
plt.show()
```

```python
x = np.array([1,4,5,7,8,10,20])
y = np.array([1,4,5,17,11,2,14])
plt.plot(x,y,marker='o')
plt.show()
```

```python
plt.plot(x,y,'*:r')
```

```python
plt.plot(x,y,'*:g',ms =20)
```

```python
plt.plot(x,y,'*:g',ms =20,mec = 'r')
```

```python
plt.plot(x,y,'*:g',ms =20, mec ='r',linewidth =5)
```

```python
plt.plot(x,y,'*:g',ms =20, mec ='r',linewidth = 2,linestyle = '-')
```

```python
plt.xlabel('number')
plt.ylabel('age')
plt.title('relashtion between age and number')
plt.plot(x,y,'*:g',ms =20, mec ='r',linewidth = 2)
```

```python
plt.title('relashtion between age and number',color = 'r',loc = 'left')
```

```python
plt.ylabel('age',rotation = 90)
plt.xlabel('number', rotation = 90)
```

```python
plt.scatter(x,y,color = 'g', marker = '<')
```
